```python
df=df.drop_duplicates()
#removing duplicate values
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
print(df['workclass'].value_counts())
print(df['occupation'].value_counts())
print(df['native-country'].value_counts())
```

```python
df = df.replace('?', np.nan)
print(df.isna().sum())
```

```python
for col in ['workclass', 'occupation', 'native-country']:
    df[col] = df[col].fillna(df[col].mode()[0])
    #replacing null values by mode value
```

```python
df.isna().sum()
#All Null value is now replaced with mode value
```

```python
df.workclass.unique()
#To double check is actually null value is replaced now or not
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
for col in [col for col in df.columns if df[col].dtype == "object"]:
    df[col] = label.fit_transform(df[col])
    #models receives only int as input not string.
    #so we converted str values to int.
```

```python
df
```

```python
df.dtypes
#cross cheking whether all objective value has been replaced with int or not
```

```python
y=df['salary']
```

```python
y
```

```python
X=df.drop('salary',axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
"""
The purpose of this code is to calculate the accuracy of a machine learning model's predictions.
By printing the accuracy score,
you can get a sense of how well your model is doing and make informed decisions about further improvements.
"""
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
print(accuracy_score(y_test,y_pred))
"""
This code snippet is typically used to assess the performance of a classification model
by comparing its predictions (y_pred) with the actual labels (y_test).
 The accuracy score provides a measure of how well the model is performing.
"""
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test,y_pred))

"""
This code does the following:

1. Trains a Random Forest Classifier on X_train and y_train.
2. Uses the trained model to make predictions on X_test.
3. Calculates the accuracy of the model by comparing the predicted labels (y_pred) with the actual labels (y_test).
4. Prints the accuracy of the model.
"""
```

```python
# Import the Gaussian Naive Bayes classifier from scikit-learn
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
"""
The purpose of this code is to:

1. Train a Gaussian Naive Bayes classifier on the training data (X_train and y_train).
2. Use the trained classifier to make predictions on the test data (X_test).
3. Evaluate the performance of the classifier by calculating its accuracy using the accuracy_score function from scikit-learn.

The code snippet is comparing the performance of two different models, as indicated by the output of another model's accuracy score (0.8603257529194838) above the code.

The accuracy of the Gaussian Naive Bayes classifier is approximately 80.21% (0.8020897357098955).
"""
```
